# Calculating the Bilingual Evaluation Understudy (BLEU) score

In this lab, we will implement a popular metric for evaluating the quality of machine-translated text: the BLEU score proposed by Kishore Papineni, et al. in their 2002 paper ["BLEU: a Method for Automatic Evaluation of Machine Translation"](https://www.aclweb.org/anthology/P02-1040.pdf), the BLEU score works by comparing "candidate" text to one or more "reference" translations. The result is better the closer the score is to 1. Let's see how to get this value in the following sections.

## 1.2  Defining the BLEU Score

You have seen the a way for calculating the BLEU score in this week's lectures. More formally, we can express the BLEU score as:

$$BLEU = BP\times\Bigl(\prod_{i=1}^{4}precision_i\Bigr)^{(1/4)} \tag{1}$$


We used the productory and the denominator of the exponent as 4 because we are assuming that we are going to work up to 4-grams. The Brevity Penalty defined as an exponential decay:

$$BP = min\Bigl(1, e^{(1-({ref}/{cand}))}\Bigr) \tag{2}$$

In the previous equation, ${ref}$ and ${cand}$ refers to the length or count of words in the reference and candidate translations. The brevity penaly $BP$ helps to handle very short translations. 

The precision is defined as :

$$precision_i = \frac {\sum_{snt \in{cand}}\sum_{i\in{snt}}min\Bigl(m^{i}_{cand}, m^{i}_{ref}\Bigr)}{w^{i}_{t}} \tag{3}$$

where:

* $m^{i}_{cand}$, is the count of i-gram in candidate matching the reference translation.
* $m^{i}_{ref}$, is the count of i-gram in the reference translation.
* $w^{i}_{t}$, is the total number of i-grams in candidate translation.


## 1.3 Explaining the BLEU score

### Brevity Penalty (example):

````python
ref_length = np.ones(100)
can_length = np.linspace(1.5, 0.5, 100)
x = ref_length / can_length
y = 1 - x
y = np.exp(y)
y = np.minimum(np.ones(y.shape), y)

# Code for in order to make the plot
fig, ax = plt.subplots(1)
lines = ax.plot(x, y)
ax.set(
    xlabel="Ratio of the length of the reference to the candidate text",
    ylabel="Brevity Penalty",
)
plt.show()
````

The brevity penalty penalizes generated translations that are too short compared to the closest reference length with an exponential decay. The brevity penalty compensates for the fact that the BLEU score has no recall term.

Here's a step-by-step explanation:

1. **Setting up the lengths**:
    - `ref_length = np.ones(100)`: Creates an array of 100 ones, representing a constant reference length.
    - `can_length = np.linspace(1.5, 0.5, 100)`: Creates an array of 100 values, linearly spaced between 1.5 and 0.5. This represents varying candidate translation lengths.
    

2. **Computing the ratio and the Brevity Penalty**:
    - `x = ref_length / can_length`: Calculates the ratio of reference length to candidate length for each value in the arrays.
    - `y = 1 - x`: Computes the difference between 1 and the ratios. This will be used to calculate the brevity penalty.
    - `y = np.exp(y)`: Applies the exponential function to each element of `y`, effectively calculating the Brevity Penalty using the formula `exp(1-ratio)`.
    - `y = np.minimum(np.ones(y.shape), y)`: Ensures that the values of `y` are capped at 1, as the Brevity Penalty should not exceed 1.

When executed, this code would generate a plot showing how the Brevity Penalty changes based on the ratio of reference to candidate translation lengths. It visually represents how shorter translations (compared to the reference) receive a penalty, which decreases as the candidate length approaches the reference length.

The n-gram precision counts how many unigrams, bigrams, trigrams, and four-grams (i=1,...,4) match their n-gram counterpart in the reference translations. This term acts as a precision metric. Unigrams account for adequacy while longer n-grams account for fluency of the translation. To avoid overcounting, the n-gram counts are clipped to the maximal n-gram count occurring in the reference ($m_{n}^{ref}$). Typically precision shows exponential decay with the degree of the n-gram.

When the n-gram precision is multiplied by the BP, then the exponential decay of n-grams is almost fully compensated. The BLEU score corresponds to a geometric average of this modified n-gram precision.

## 1.4 Example Calculations of the BLEU score

In this example we will have a reference translation and 2 candidates translations. We will tokenize all sentences using the NLTK package introduced in Course 2 of this NLP specialization.

````python
# reference sentence that is the ground truth.
reference = "The NASA Opportunity rover is battling a massive dust storm on planet Mars." 

# candidate sentences that are alternative sentences that might be similar to the reference sentence.
candidate_1 = "The Opportunity rover is combating a big sandstorm on planet Mars."
candidate_2 = "A NASA rover is fighting a massive storm on planet Mars."

tokenized_ref = nltk.word_tokenize(reference.lower()) # tokenize the reference sentence.
tokenized_cand_1 = nltk.word_tokenize(candidate_1.lower()) # tokenize the first candidate sentence.
tokenized_cand_2 = nltk.word_tokenize(candidate_2.lower()) # tokenize the second candidate sentence.

print(f"{reference} -> {tokenized_ref}")
print("\n")
print(f"{candidate_1} -> {tokenized_cand_1}")
print("\n")
print(f"{candidate_2} -> {tokenized_cand_2}")
````

### STEP 1: Computing the Brevity Penalty

````python
def brevity_penalty(candidate, reference):
    ref_length = len(reference)
    can_length = len(candidate)

    # Brevity Penalty
    if ref_length < can_length: # if reference length is less than candidate length
        BP = 1 # set BP = 1
    else:
        penalty = 1 - (ref_length / can_length) # else set BP=exp(1-(ref_length/can_length))
        BP = np.exp(penalty)

    return BP
````

This line of code is based on equation 2 which calculates Brevity Penalty $BP$ which ensures translations aren't unfairly short.
- ref_length --> $\text{ref}$: The length of the reference translation.
- cand_length --> $text{cand}$: The length of the candidate translation.
- IF: the candidate translation is longer than or equal to the reference (`ref_length <= can_length`)
    - THEN: The penalty is 1 $BP = 1$ (no penalty)
- IF: the candidate is shorter (`ref_length > can_length`)
    - THEN: the penalty is an exponentially decaying function of the length ratio ($BP=e^{(1-({ref}/{cand}))}$).  

### STEP 2: Computing the Precision

````python
def clipped_precision(candidate, reference):
    """
    Clipped precision function given a original and a machine translated sentences
    """

    clipped_precision_score = []
    
    for i in range(1, 5):
        '''
        The loop iterates through i from 1 to 4, indicating it's calculating n-grams 
        from 1-gram (individual words) to 4-grams (four consecutive words).
        '''
        ref_n_gram = Counter(ngrams(reference,i)) # count the n-gram in the reference text
        cand_n_gram = Counter(ngrams(candidate,i)) # count the n-gram in the candidate text

        c = sum(cand_n_gram.values())   
        
        '''
        for sentence "I love Machine Learning" the ngrams and c would be:
        1-gram: ['I', 'love', 'Machine', 'Learning'] 4
        2-gram: ['I', 'love'], ['love', 'Machine'], ['Machine', 'Learning'] 3
        3-gram: ['I', 'love', 'Machine'], ['love', 'Machine', 'Learning'] 2
        4-gram: ['I', 'love', 'Machine', 'Learning'] 1
        '''   
        
        
        '''
        The following code handles the condition when the n-gram in the candidate text is not in the reference text.
        This loop, iterated for all n-grams, modifies the counts in cand_n_gram based on their presence and counts 
        in ref_n_gram, which will be used to compute the clipped precision as part of the BLEU score calculation 
        in the function provided earlier. The "clipped" in clipped precision refers to the minimizing of the n-gram 
        counts to the level of the reference to avoid over-crediting repeated phrases.
        '''
        for j in cand_n_gram: # for every n-gram up to 4 in candidate text
            if j in ref_n_gram: # check if it is in the reference n-gram
                # if the count of the candidate n-gram is bigger
                # than the corresponding count in the reference n-gram,
                if cand_n_gram[j] > ref_n_gram[j]: 
                    # then set the count of the candidate n-gram to be equal
                    # to the reference n-gram
                    cand_n_gram[j] = ref_n_gram[j] 
            else:
                cand_n_gram[j] = 0 # else set the candidate n-gram equal to zero

                        
        clipped_precision_score.append(sum(cand_n_gram.values())/c)
        
    weights =[0.25]*4

    s = (w_i * math.log(p_i) for w_i, p_i in zip(weights, clipped_precision_score))

    s = math.exp(math.fsum(s))
    return s
````

Based on the equation 3: 

- $\text{precision}_i$: The modified n-gram precision score for n-grams of size $i$.
    ```python
    clipped_precision_score.append(sum(cand_n_gram.values())/c)
- $\sum_{\text{snt} \in \text{cand}}\sum_{i \in \text{snt}}$: These nested sums iterate over each sentence $\text{snt}$ in the candidate translation and over each n-gram $i$ of size $i$ in those sentences.
    ```python
    sum(cand_n_gram.values())
- $m^{i}_{\text{cand}}$: The count of n-gram $i$ in the candidate translation.
    ```python
    cand_n_gram = Counter(ngrams(candidate,i))
- $m^{i}_{\text{ref}}$: The count of n-gram $i$ in the reference translation.
    ```python
    ref_n_gram = Counter(ngrams(reference,i))
- $\min \left( m^{i}_{\text{cand}}, m^{i}_{\text{ref}} \right)$: The clipped count of n-gram $i$, which ensures that we donât over-count matching n-grams. </br>
    ```python
    if cand_n_gram[j] > ref_n_gram[j]: 
        cand_n_gram[j] = ref_n_gram[j]`
- $w^{i}_{t}$: The total number of n-grams of size $i$ in the candidate translation.
    ```python
    c = sum(cand_n_gram.values())   
- So, $\text{precision}_i$ is calculated as the ratio of the clipped count of n-grams to the total count of n-grams, summed over all sentences and all n-grams.

$$precision_i = \frac {\sum_{snt \in{cand}}\sum_{i\in{snt}}min\Bigl(m^{i}_{cand}, m^{i}_{ref}\Bigr)}{w^{i}_{t}} \tag{3}$$

### STEP 3: Computing the BLEU score

````python
def bleu_score(candidate, reference):
    BP = brevity_penalty(candidate, reference)    
    precision = clipped_precision(candidate, reference)    
    return BP * precision
````

This last function corresponds to equation 1: 

- $\text{BLEU}$: The final BLEU score, a metric used to evaluate machine translations. Ranges from 0 to 1 (or 0% to 100%). Higher scores indicate better translations.
- $BP$: Brevity Penalty. Ensures that shorter translations don't get unfairly high scores.
- $\text{precision}_i$: Modified n-gram precision for n-grams of size $i$. Calculated for $i$ from 1 to 4 (unigrams, bigrams, trigrams, and 4-grams).
- The product $\prod_{i=1}^{4}\text{precision}_i$ represents the geometric mean of the 4 precision scores, which is then power-rooted by $1/4$ to get the final mean used in the BLEU score calculation.


### STEP 4: Testing with our Example Reference and Candidates Sentences

````python
print(
    "Results reference versus candidate 1 our own code BLEU: ",
    round(bleu_score(tokenized_cand_1, tokenized_ref) * 100, 1),
)

print(
    "Results reference versus candidate 2 our own code BLEU: ",
    round(bleu_score(tokenized_cand_2, tokenized_ref) * 100, 1),
)
````

### STEP 5: Comparing the Results from our Code with the Sacrebleu Library

````python
print(
    "Results reference versus candidate 1 sacrebleu library sentence BLEU: ",
    round(sacrebleu.sentence_bleu(candidate_1, [reference]).score, 1),
)
print(
    "Results reference versus candidate 2 sacrebleu library sentence BLEU: ",
    round(sacrebleu.sentence_bleu(candidate_2, [reference]).score, 1),
)
````

In this section, we will show a simple pipeline for evaluating machine translated text. Instead, we will be using [Google Translate](https://translate.google.com) to generate English to German translations and we will evaluate it against a known evaluation set. There are three files we will need:

1. A source text in English. In this lab, we will use the first 1671 words of the [wmt19](http://statmt.org/wmt19/translation-task.html) evaluation dataset downloaded via SacreBLEU. 2. A reference translation to German of the corresponding first 1671 words from the original English text. This is also provided by SacreBLEU.
3. A candidate machine translation to German from the same 1671 words. This is generated by feeding the source text to a machine translation model. As mentioned above, we will use Google Translate to generate the translations in this file.

With that, we can now compare the reference and candidate translation to get the BLEU Score.

````python
# Loading the raw data
wmt19_src = open("data/wmt19_src.txt", "r")
wmt19_src_1 = wmt19_src.read()
wmt19_src.close()

wmt19_ref = open("data/wmt19_ref.txt", "r")
wmt19_ref_1 = wmt19_ref.read()
wmt19_ref.close()

wmt19_can = open("data/wmt19_can.txt", "r")
wmt19_can_1 = wmt19_can.read()
wmt19_can.close()

tokenized_corpus_src = nltk.word_tokenize(wmt19_src_1.lower())
tokenized_corpus_ref = nltk.word_tokenize(wmt19_ref_1.lower())
tokenized_corpus_cand = nltk.word_tokenize(wmt19_can_1.lower())
````
